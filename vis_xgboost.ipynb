```python
import seaborn as sns
import pandas as pd

# These lines set up the plotting funstionality and formatting.
import matplotlib 
matplotlib.use('Agg', warn = False)
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

import numpy as np

```

```python
#X_test = pd.read_pickle("X_test.pickle")
df= pd.read_pickle("graph_train.pickle")
```

```python
df2 = pd.read_pickle("graph_test.pickle")
```

```python
df2.head()
```

```python
plt.figure(figsize=(15,15))
plt.hlines(y=0, xmin=-10, xmax=350000,color='black', lw=0.8)
sns.scatterplot(x = "sub",y = "y_train_pred",data = df,alpha=0.8, color = "steelblue",marker = "o",edgecolor="white", label = "Training Data")
sns.scatterplot(x = "sub", y= "y_test_pred", data = df2,alpha=0.4,color = "limegreen",marker = 's', edgecolor = "white",
               label = "Test Data")
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.savefig('Residual_plot_Xgboost.png', bbox_inches='tight',dpi=800)

plt.show()
```
